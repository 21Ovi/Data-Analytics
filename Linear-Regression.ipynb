# Data Preprocessing

```python
df.isnull().sum()
```

```python
df.head()
```

```python
box=df[['age','bmi']]
plt.figure(figsize=(5,10),facecolor='white')
plotnumber=1

for column in box:
    if plotnumber <=3:
        ax = plt.subplot(3,1,plotnumber)
        sns.boxplot(x=box[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
```

* Methods to handle categorical data
 <br> 1 using loc function and domain analysis
 <br> 2 One hot encoding
 
 <br> sudo code for loc function:
 <br> df.loc[df[column] == '',column] = 'some numerical value'
